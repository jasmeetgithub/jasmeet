```python
#dropping unwanted column

df.drop(['event_time','product_id','category_id','user_session','price','user_id','event_type'],axis=1,inplace=True)
df.head(2)
```

```python
#Replacing special Character

category_codes = df['category_code'].tolist()
category_codes = [code.replace('.', ',').replace('_', ' ') for code in category_codes]
df['category_code'] = category_codes
```

```python
# We need only unique values in brand for model.

df = pd.DataFrame(df)

unique_brands = df.drop_duplicates(subset='brand')

new_df = pd.DataFrame(unique_brands, columns=['brand', 'category_code'])
df.update(new_df)
print(new_df)
```

```python
new_df_2=pd.read_csv('Untitled87.csv')
```

```python
new_df_2.shape
```

```python
new_df_2.head(29)
```

```python
#model

from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=500)
```

```python
cv
```

```python
#Training the data

vectors=cv.fit_transform(new_df_2['category_code']).toarray()
```

```python
vectors
```

```python
#for checking score in category, so as you can see its matching perfectly by value.

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
#final function to reccomend

def recommend(brand):
    company_name=new_df_2[new_df_2['brand'] == brand].index[0]
    distances=similarity[company_name]
    name_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:40]
    
    for i in name_list:
        print(new_df_2.iloc[i[0]].brand)
```

```python
recommend('samsung')
```
